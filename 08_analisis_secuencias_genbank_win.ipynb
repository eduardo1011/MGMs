```python
#funcion para contar en una secuencia cuantas bases de cada una hay, detectando n y otras bases
def contar_n(sec2):
    a, c, g, t, n, o, n1 =0, 0, 0, 0, 0, 0, 0 # n1 es un contador
    for s in sec2:
        if s=="A": ## == significa comparar
            a+=1
        elif s=="C":  # elif condicionantes anidadas 
            c+=1
        elif s=="G":
            g+=1
        elif s== "T":
            t+=1
        elif s== "N":
            n+=1
        else:
            o+=1
            print (s, end=",")
        #print (s, end=",")
        n1+=1
        if n1%10000000==0:
            #break
            print ("procesando", n1)
    print ("a=", a, "\t","c=", c, "\t","g=", g, "\t","t=", t, "\t","n=", n,"o=", "\t", o)
    print (a+c+g+t, "nucleotidos", n,"("+str( round(n/len (sec2)*100,4))+"%)","\n",
           o,"("+str( round(o/len (sec2)*100,4))+"%)")
    return (a,c,g,t,n,o)
```

```python
A1,C1,G1,T1,N1,O1 = contar_n(seq_record.seq)
print("A =", A1, "C =", C1, "G =", G1, "T =", T1, "N =", N1, "otros =", O1)
print ("GC (%)", (C1+G1)/(A1+C1+G1+T1+N1+O1)*100, "%" )
```

```python
A1,C1,G1,T1,N1,O1 = contar_n(seq_record.seq)
print("A =", A1, "C =", C1, "G =", G1, "T =", T1, "N =", N1, "otros =", O1)
print ("GC (%)", round((C1+G1)/(A1+C1+G1+T1+N1+O1)*100,2), "%" ) #round redondea
```

```python
#nucleotidos
secuencias= []
Entrez.email = "pgen10@hotmail.com"
handle = Entrez.efetch(db="nucleotide", rettype="gb", retmode="text", id=lineas)
for seq_record in SeqIO.parse(handle, "gb"): #parse 
    print("%s %s..." % (seq_record.id, seq_record.description[:50]))
    print("Sequence length %i, %i features, from: %s"
          % (len(seq_record), len(seq_record.features), seq_record.annotations["source"]))
    secuencias.append(seq_record)
SeqIO.write(secuencias, "argos.gb", "gb")
SeqIO.write(secuencias, "argos.fas", "fasta")

handle.close()
```
